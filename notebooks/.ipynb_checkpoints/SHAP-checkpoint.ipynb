```python
import sys
sys.path.append('..')
import os
from loltest import *
from efficientnet_pytorch.model import *
import shap
```

```python
home = '/media/ags/DATA/CODE/kaggle/aptos2019-blindness-detection/'
ckpt_path = os.path.join(home, "weights/198_efficientnet-b5_f1_ptest/ckpt20.pth")
predict_on = 'test' 
size = 256
model_name = "efficientnet-b5"
if predict_on == "test":
    sample_submission_path = os.path.join(home, "data/sample_submission.csv")
elif predict_on == "train":
    sample_submission_path =  os.path.join(home, "data/train.csv")
elif predict_on == "train_mess":
    sample_submission_path = os.path.join(home, "data/train_messidor.csv")

tta = 0  # number of augs in tta
root = os.path.join(home, "data/test_images/")
if predict_on == "train":
    root = os.path.join(home, "data/all_images/")
elif predict_on == "train_mess":
    root = os.path.join(home, "external_data/messidor/train_images/")
#root = 'data/npy_files/bgcc456'
mean = (0.485, 0.456, 0.406)
std = (0.229, 0.224, 0.225)
# mean = (0, 0, 0)
# std = (1, 1, 1)
use_cuda = False
num_classes = 1
num_workers = 2
batch_size = 2
device = torch.device("cuda" if use_cuda else "cpu")
if use_cuda:
    cudnn.benchmark = True
    torch.set_default_tensor_type("torch.cuda.FloatTensor")
else:
    torch.set_default_tensor_type("torch.FloatTensor")

best_sub = os.path.join(home, "weights/submission812.csv")
df = pd.read_csv(sample_submission_path)
testset = DataLoader(
    TestDataset(root, df, size, mean, std, predict_on, tta),
    batch_size=batch_size,
    shuffle=False,
    num_workers=num_workers,
    pin_memory=True if use_cuda else False,
)

model = get_model(model_name, num_classes, pretrained=None)
model.to(device)
model.eval()
state = torch.load(ckpt_path, map_location=lambda storage, loc: storage)
model.load_state_dict(state["state_dict"])
```

```python
batch = next(iter(testset))
background = []
for i in range(10):
    background.append(batch[:, 0, :])
background = torch.cat(background, axis=0)
batch = next(iter(testset))
test_images = batch[:, 0, :]
```

```python
background.shape, test_images.shape
```

```python
e = shap.DeepExplainer(model, background)
```

```python
shap_values = e.shap_values(test_images)
```

```python
def unnormalize(img):
    img = img.clone() # V. Imp
    mean = torch.as_tensor([0.485, 0.456, 0.406])
    std = torch.as_tensor([0.229, 0.224, 0.225])
    #return (img * std) + mean
    for t, m, s in zip(img, mean, std):    
        t.mul_(s).add_(m)
    #img = (img * 255)
    return img
```

```python
un_test_images = [unnormalize(img).numpy() for img in test_images]
```
